## Step 1: Extract data from normalized SQL schema


```python
from  flask import Flask,request,jsonify
import pandas as pd
import numpy as np
from dotenv import load_dotenv
import os
import pyodbc
```

```python
app=Flask(__name__)
```

```python


load_dotenv()

def get_db_connection():

    conn_str = (
        f"DRIVER={{{os.getenv('DB_DRIVER')}}};"
        f"SERVER={os.getenv('DB_SERVER')};"
        f"DATABASE={os.getenv('DB_NAME')};"
        f"Trusted_Connection={os.getenv('DB_TRUSTED_CONNECTION')};"
    )

    return pyodbc.connect(conn_str)


def insert_csv_to_table(csv_path, table_name):
    conn = get_db_connection()
    cursor = conn.cursor()

    try:
        df = pd.read_csv(csv_path)
        print(f" data is being entered from üìÇ  {csv_path} to table{table_name} ...")

        for index, row in df.iterrows():
            placeholders = ','.join(['?'] * len(row))
            columns = ','.join(df.columns)
            sql = f"INSERT INTO {table_name} ({columns}) VALUES ({placeholders})"
            cursor.execute(sql, tuple(row))

        conn.commit()
        print(f" data entered ‚úÖ {len(df)}  {table_name}.\n")
    except Exception as e:
        print(f"‚ùå error  while entering files{table_name}: {e}")
        conn.rollback()

    finally:
        cursor.close()
        conn.close()

```

```python
insert_csv_to_table(r"Data\customers_data_10k.csv", "Customers")
```
